## Categorical Plots
Distributions with reference to numerical data

### Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator= np.var) # estimator could be any aggregation function np.mean np.var np.std
```

### Count Plots

```python
sns.countplot(x='sex', data=tips_df)
```

### Box Plots
Shows median, std and outliers of data<br>
[ELI5](https://www.simplypsychology.org/boxplots.html)

```python
sns.boxplot(x="day", y="total_bill", data=tips_df, hue="sex") # hue is used to add a second variable to compare
```

### Violin Plot
A box plot with KDE surrounding it. Shows the estimated smoothed distribution of the sample estimated over the entire population.

```python
sns.violinplot(x="day", y="total_bill", data=tips_df, hue="sex", split=True) # combines the hue subcategory into one plot
```

### Strip Plot
Draws a scatter graph of the data where one variable is categorical

```python
plt.figure(figsize=(8, 5))

sns.stripplot(x="day", y="total_bill", data=tips_df, jitter=True, hue="sex", dodge=True) # jitter stretches the data points out, dodge seperates the hues out
```

### Swarm plot
A strip plot of a violin plot

```python
sns.violinplot(x="day", y="total_bill", data=tips_df)
sns.swarmplot(x="day", y="total_bill", data=tips_df, color="white")
```

### Pair Grids
Pair grid similar to pair plot but gives more control over layout

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue="species")
# iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist) # controls diagonal plot
# iris_g.map_offdiag(plt.scatter) # controls off the diagonal
iris_g.map_lower(sns.kdeplot) # controls off diagonal lower
iris_g.map_upper(plt.scatter) # controls off diagonal upper

```

```python
# define x.y axis
iris_g = sns.PairGrid(iris, 
                      hue="species", 
                      x_vars=["sepal_length", "sepal_width"],
                      y_vars=["petal_length", "petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

### Facet Grid
Facet Grid will plot multiple aggregate function plots split by univariate data. The below we split and plot the data by time of meal and smoker or not

```python
tips_fg = sns.FacetGrid(tips_df, col="time", row="smoker")
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_fg = None # clears the variable 
tips_fg = sns.FacetGrid(tips_df, 
                        col="time", # column of the grid
                        row="smoker") # row data of the drid
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = None # clears the variable 
tips_fg = sns.FacetGrid(tips_df, 
                        col="time", # column
                        hue="sex", # data categorisation
                        height=4, 
                        aspect=1.3,
                        col_order= ["Dinner", "Lunch"], #order of the graphs
                        palette="Set2")
tips_fg.map(sns.scatterplot, 
            'total_bill', # 
            'tip', 
            edgecolor="b")
tips_fg.add_legend()
```

```python
kws = dict(s= 50,
           linewidth= .5,
           edgecolor="w")
tips_fg = sns.FacetGrid(tips_df, 
                    col="sex", # column
                    hue="smoker", # data categorisation
                    height=4, 
                    aspect=1.3,
                    hue_kws=dict(marker=["^", "v"]))
tips_fg.map(plt.scatter,
            "total_bill",
            "tip",
            **kws)
tips_fg.add_legend()
```
